```python
from sklearn.datasets import load_iris
from sklearn.datasets import make_moons

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix

iris = load_iris()
x_data = iris.data[:,:]
y_data = iris.target
```

```python
_ = scatter_matrix(pd.DataFrame(x_data), alpha=0.7,
                   c=y_data, s=100, figsize=(12,12), diagonal="kde")
```

```python
pca_result = PCA(2).fit_transform(x_data)

plt.scatter(pca_result[:,0], pca_result[:,1], c = iris.target)
```

```python
tsne_result = TSNE(n_components=2, init='pca').fit_transform(x_data)

plt.scatter(tsne_result[:,0], tsne_result[:,1], c = iris.target)
```

-----------

```python
x_moon, y_moon = make_moons(n_samples=1000, shuffle=True, noise=0.05)
```

```python
plt.scatter(x_moon[y_moon == 0, 0], x_moon[y_moon == 0, 1], label="Class 0", alpha=0.7)
plt.scatter(x_moon[y_moon == 1, 0], x_moon[y_moon == 1, 1], label="Class 1", alpha=0.7)
plt.legend()
```

```python
pca_result = PCA(1).fit_transform(x_moon)
```

```python
plt.figure(figsize=(8,8))
plt.subplot(211)
plt.title("PCA")
plt.scatter(pca_result[y_moon == 0, 0], np.zeros(500), label="Class 0", alpha=0.5)
plt.scatter(pca_result[y_moon == 1, 0], np.zeros(500), label="Class 1", alpha=0.5)
plt.legend()
plt.subplot(212)
plt.title("PCA")
plt.scatter(pca_result[y_moon == 0, 0], np.zeros(500), label="Class 0", alpha=0.5)
plt.scatter(pca_result[y_moon == 1, 0], np.ones(500), label="Class 1", alpha=0.5)
plt.legend()
```

----------

```python
tsne_result = TSNE(n_components=1, init='pca').fit_transform(x_moon)

plt.figure(figsize=(8,8))
plt.subplot(211)
plt.title("tSNE")
plt.scatter(tsne_result[y_moon == 0, 0], np.zeros(500), label="Class 0", alpha=0.5)
plt.scatter(tsne_result[y_moon == 1, 0], np.zeros(500), label="Class 1", alpha=0.5)
plt.legend()
plt.subplot(212)
plt.title("tSNE")
plt.scatter(tsne_result[y_moon == 0, 0], np.zeros(500), label="Class 0", alpha=0.5)
plt.scatter(tsne_result[y_moon == 1, 0], np.ones(500), label="Class 1", alpha=0.5)
plt.legend()
```
